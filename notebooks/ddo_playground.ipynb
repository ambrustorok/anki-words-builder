```python
def make_url(query:str):
    return f"https://ordnet.dk/ddo/ordbog?query={query}"
```

```python
import requests
from bs4 import BeautifulSoup

def get_artikel_div(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find the div with class "artikel"
        artikel_div = soup.find('div', class_='artikel')
        
        if artikel_div:
            return artikel_div
        else:
            return "No div with class 'artikel' found on the page."
    else:
        return f"Failed to retrieve the webpage. Status code: {response.status_code}"
```

```python
# Example usage
url = make_url("hund")
result = get_artikel_div(url)
print(result)
```

```python
import requests
from bs4 import BeautifulSoup

def get_artikel_content(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find the div with class "artikel"
        artikel_div = soup.find('div', class_='artikel')
        
        if artikel_div:
            # Find all definitionBox divs within artikel
            definition_boxes = artikel_div.find_all('div', class_='definitionBox')
            
            if len(definition_boxes) >= 2:
                # Keep everything up to and including the second definitionBox
                content = ''
                for child in artikel_div.children:
                    content += str(child)
                    if child == definition_boxes[1]:
                        break
                
                return content.strip()
            else:
                return "Couldn't find two 'definitionBox' divs within 'artikel'."
        else:
            return "No div with class 'artikel' found on the page."
    else:
        return f"Failed to retrieve the webpage. Status code: {response.status_code}"

# Example usage
url = make_url("hund")
result = get_artikel_content(url)
print(result)
```
